```python
import matplotlib.pyplot as plt
```

```python
#라인 그래프 출력
plt.plot([1,2,3,4])
plt.show
```

```python
#라인 그래프 x 축의 값 y 축의 값지정
plt.plot([1,2,3,4], [1,4,2,8])
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,8]

plt.plot(x,y,"--r")  #-- : 점선 / r : 빨간색
plt.plot(x, [4,2,6,4], "--b") # o 는 점 / y : 노락색
plt.xlabel("Class") #x축
plt.ylabel("Count") #y축
plt.title("Class & Count") # 제목
plt.legend(["A","B"], loc=3) # 범주
plt.show
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot([1,2,3,4])
plt.subplot(1,2,2)
plt.plot([8,7,6,5])
plt.show
```

1. 임의의 그래프 4개를 2x2 subplot 출력

```python
plt.subplot(2,2,1)
plt.plot([1,2,3,4])
plt.subplot(2,2,2)
plt.plot([4,3,2,1])
plt.subplot(2,2,3)
plt.plot([4,3,2,1])
plt.subplot(2,2,4)
plt.bar([1,2,3,4],[10,5,3,2])
plt.show
```

```python
#막대그래프
x = [1,2,3,4]
y = [80,30,40, 60]
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,color=["r", "b"], width = 0.4)
plt.xticks(x, ["maths", "science", "history", "english"])
plt.show()
```

```python
import numpy as np
```

```python
#산점도 그래프 그리기
x = np.random.rand(50)
y = np.random.rand(50)
print(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#산점도 그래프 옵션 추가 (영역 확장, 색상 변경, 투명도 조절)

color = np.random.rand(50)
area = (20 * np.random.rand(50))**2

plt.scatter(x,y, s=area, c=color, alpha= 0.3) #alpha는 투명도
plt.show()
```
