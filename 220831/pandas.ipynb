## 파생변수 생성
1. 원가, 단가, 판매단위 값을 기준으로 
2. 총수익, 총비용, 총이윤 새로운 컬럼을 생성
3. 총수익 = 단가 * 판매단위
4. 총비용 = 원가 * 판매단위
5. 총이윤 = 총수익 - 총비용

```python
##새로운 컬럼을 추가하는 방법
##데이터 프레임명 [새로운 컬럼의 이름] =  컬럼에 들어갈 값(Series, list)
##단가 컬럼 * 판매단위 컬럼
sales_df_copy2['총수익'] = sales_df_copy2["Unit Price"] * sales_df_copy2["Units Sold"]
```

```python
## 총비용, 총이윤
sales_df_copy2["총비용"] = sales_df_copy2["Unit Cost"] * sales_df_copy2["Units Sold"]
sales_df_copy2["총이윤"] = sales_df_copy2["총수익"] - sales_df_copy2["총비용"]
sales_df_copy2.head()
```

1. corona.csv 로드 데이터 프레임화
2. 데이터를 복사해서 새로운 변수
3. 결측치가 존재하는지 1차적으로 확인
4. 컬럼의 이름 변경 --> 등록일시, 사망자, 확진자, 게시글번호, 기준일, 기준시간, 수정일시, 누적의심자, 누적확진률
5. 결측치가 존재하는 컬럼을 삭제 (현실에서 삭제는 비추)
6. 등록일시를 기준으로 오름차순 정렬
7. 인덱스의 값을 재지정 (기존의 인덱스 값은 삭제)
8. 등록일시 데이터의 형태를 datetime 변경

```python
test = pd.read_csv("corona.csv")
test.drop(["Unnamed: 0"], axis = "columns", inplace = True)
test
```

```python
##결측치 확인
test.isna()
test.isna().sum()

## 컬럼의 이름을 변경
test.columns = ['등록일시', '사망자', '확진자', '게시글번호', '기준일', '기준시간', '수정일시', '누적의심자', '누적확진률']
test.head()
```

```python
#결측치가 존재하는 컬럼을 삭제 
test.dropna(axis = "columns", inplace=True)
test.head()
```

```python
#6. 등록일시를 기준으로 오름차순 정렬
test.sort_values(["등록일시"], inplace = True)



#7. 인덱스의 값을 재지정 (기존의 인덱스 값은 삭제)
test = test.reset_index(drop = True)


#8. 등록일시 데이터의 형태를 datetime 변경
test["등록일시"] = pd.to_datetime(test["등록일시"], format = "%Y-%m-%d %H:%M:%S")
test.head()
```

## shift()
1. shift(n) --> n을 공란으로 두게 되면 기본 값이 1
2. n 값만큼 인덱스를 이동하여 그 행의 값을 가지게 된다.

### diff()
1. 한 객체 내에서 열과 열 / 행과 행 차이를 출력

```python
# shift() 함수를 이용하는 방법
test["확진자"] - test["확진자"].shift()
```

```python
# diff()
test["확진자"].diff()
```

```python
##일일 확진자, 일일 사망자라는 컬럼을 추가
test["일일확진자"] = test["확진자"] - test["확진자"].shift()
test["일일사망자"] = test["사망자"].diff()
test.head()
```
