### concat() 
1. 데이터프레임을 결합하는 함수
2. axis 0,1 행을 추가, 열을 추가 지정(기본값은 행을 추가)
3. 단순하게 결합
4. ignore_index 속성 기본값이 false 인덱스의 값은 그대로 (True) 할 것인가 아니면 새로 재지정 (False)


```python
#pd.concat([df,df2], axis = 1)  #'rows = 0' "columns = 1"

pd.concat([df,df2], axis = 'rows', ignore_index= True)  #'rows' "columns"

#pd.concat([df,df2], axis = 'rows').reset_index(drop=True)   위랑 같은 의미
```

```python
pd.concat([df,df2], axis = 'rows', ignore_index= False) 
```

```python
pd.concat([df,df2], axis = 'columns', ignore_index= True) 
```

```python
pd.concat([df,df2], axis = 'columns', ignore_index= False) 
```

## merge()
1. 데이터프레임을 결합 함수
2. 특정 조건을 만족해야 열이 추가
3. on 속성 --> 조건 
4. how 속성 --> 기준이 되는 데이터프레임 선택


```python
df4 = pd.DataFrame({
    "id" : [1,2,3,4,5], 
    "item" : ["a", 'b', 'a', 'a', 'b']
})
df5 = pd.DataFrame({
    "id" : [1,4,5,2,3],
    "name" : ["kim", "lee", "park", 'choi', 'ha']
})
```

```python
df4
```

```python
df5
```

```python
#concat() 열 추가
#merge() 열 추가
pd.concat([df4, df5], axis= 'columns')

#pd.concat([df4, df5], axis= 1)
```

```python
pd.merge(df4, df5, on='id', how='inner')#inner 는 겹치는 /right / left #id 값에 따라 name 값의 순서도 바뀜
```

```python
pd.merge(df4, df5, on='id', how='right')#inner 는 겹치는 /right / left #id 값에 따라 name 값의 순서도 바뀜
```

```python
pd.merge(df4, df5, on='id', how='left')#inner 는 겹치는 /right / left #id 값에 따라 name 값의 순서도 바뀜
```

```python
df6 = pd.DataFrame({
    "item" : ['a', 'b', 'c'],
    "item_price" : [100, 500, 1000]
})
```

```python
pd.concat([df4, df6], axis=1)
```

```python
pd.merge(df4, df6, on='item', how="inner")
```

```python
pd.merge(df4, df6, on='item', how="outer")
```
